Now are data is ready for Machine Learning

```python
x=data.drop('Item_Outlet_Sales',axis=1)
x
```

```python
y=data.loc[:,'Item_Outlet_Sales'].values.reshape(-1,1)
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
max_r2_score=0
for r_state in range(42,101):
    train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=r_state,test_size=0.3)
    lr=LinearRegression()
    lr.fit(train_x,train_y)
    pred=lr.predict(test_x)
    r2_scr=r2_score(pred,test_y)
    if r2_scr>max_r2_score:
        max_r2_score=r2_scr
        final_r_state=r_state
    print('Max r2 score corresponding to ',final_r_state,' is ',max_r2_score)    
```

```python
a_score=cross_val_score(lr,x,y,cv=5,scoring='r2')
a_score
```

```python
#Final model
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=80,test_size=0.30)
lr.fit(train_x,train_y)
y_pred=lr.predict(test_x)
print('Predicted_Item_Outlet_Sales',y_pred)
print('Real Sales',test_y)
```

```python
print('Mean Absolute Error',mean_absolute_error(y_pred,test_y))
print('Mean_squared_Error',mean_squared_error(y_pred,test_y))
print('Root Mean Squared Error',np.sqrt(mean_squared_error(y_pred,test_y)))
print('R2 Score',r2_score(y_pred,test_y))
```
